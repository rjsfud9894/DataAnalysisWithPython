### 1-4 pandas 이용해 CCTV와 인구 현황 데이터 파악하기

```python
CCTV_Seoul.head()
```

```python
CCTV_Seoul.sort_values(by='소계', ascending=True).head(5)
```

```python
CCTV_Seoul.sort_values(by='소계', ascending=False).head(5)
```

```python
CCTV_Seoul['최근증가율'] = (CCTV_Seoul['2016년'] + CCTV_Seoul['2015년'] 
                        + CCTV_Seoul['2014년']) / CCTV_Seoul['2013년도 이전'] * 100
```

```python
CCTV_Seoul.sort_values(by='최근증가율', ascending=False).head(5)
```

```python
pop_Seoul.head()
```

```python
pop_Seoul.drop([0], inplace=True) # drop 명령을 사용해 현재 필요 없는 합계 필드를 삭제한다.
pop_Seoul.head()
```

```python
pop_Seoul['구별'].unique() # '구별' 컬럼의 한번 이상 나타난 데이터를 확인 함
```

```python
pop_Seoul[pop_Seoul['구별'].isnull()] # isnull 명령을 이용해 NaN 데이터를 추출한다.
```

```python
pop_Seoul.drop([26], inplace=True) # drop 명령으로 NaN이 있던 행을 삭제한다.
pop_Seoul.head()
```

```python
pop_Seoul['외국인 비율'] = pop_Seoul['외국인'] / pop_Seoul['인구수'] * 100
pop_Seoul['고령자 비율'] = pop_Seoul['고령자'] / pop_Seoul['인구수'] * 100
pop_Seoul.head()
```

```python
pop_Seoul.sort_values(by='인구수', ascending=False).head(5)
```

```python
pop_Seoul.sort_values(by='외국인', ascending=False).head(5)
```

```python
pop_Seoul.sort_values(by='외국인 비율', ascending=False).head(5)
```

```python
pop_Seoul.sort_values(by='고령자', ascending=False).head(5)
```

```python
pop_Seoul.sort_values(by='고령자 비율', ascending=False).head(5)
```
